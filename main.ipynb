```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import torch
```

## CoinGame

```python
from coin_game import CoinGameGPU

env = CoinGameGPU(device=torch.device("cpu"))
obs_blue, obs_red = env.reset()
for t in range(100):
    actions = [torch.randint(env.NUM_ACTIONS, size=(env.batch_size,)), torch.randint(env.NUM_ACTIONS, size=(env.batch_size,))]
    env.step(actions)
    if t == 66:
        plt.imshow(env.blue_rgb_state[0].cpu().numpy().transpose(1, 2, 0))
```
